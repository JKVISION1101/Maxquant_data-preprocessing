```python
class Maxquant():
    
    def Getuniprotein(self,data,column):
        data["select_protein"]=data[column].str.rsplit(pat=';',)
        data.dropna(subset = ["select_protein"], inplace = True)
        data["select_protein"]=data["select_protein"].apply(lambda x: x[0])
        data['Accession']=""
        data["protein name"]=""
        data=data[data["select_protein"] != '']
        data.reset_index(drop=True,inplace=True)
        u=re.compile('\|(.+)\|')
        p=re.compile('ARATH (.+) OS')
        for i in range(0,data[column].shape[0]):
            if p.search(data["select_protein"][i])== None:
                pass
            else:
                u_1=u.search(data["select_protein"][i]).group(1)
                p_1=p.search(data["select_protein"][i]).group(1)
                data['Accession'][i]=u_1
                data['protein name'][i]=p_1
        return data
    
    def remove_noid(self,data):
        data_1 = data.dropna(axis = 0, subset = ['id'])
        return data_1
        
        

    def pairs(self,n,DF3=DF3.copy()):
        DF3[DF3_column]=DF3[DF3_column].str.split(pat=";")
        DF3=DF3.dropna(subset=[DF3_column])
        DF3=DF3.reset_index(drop=True)
        list=[]
        for i in range(0,DF3[DF3_column].shape[0]):
            if len(DF3[DF3_column][i]) ==n:
                list.append(i)
            else:
                pass
        return DF3.loc[list].reset_index(drop=True)
    
    def pairs_1(self,data,address):
        data_1=data.copy()
        data_1[DF3_column]=data_1[DF3_column].apply(lambda x : ''.join(x))
        data_1[DF3_column]=data_1[DF3_column].apply(lambda x: x.split("_"))
        for i in range(0,data_1.shape[0]):
            data_1[DF3_column][i].sort()
        data_1[DF3_column]=data_1[DF3_column].apply(lambda x: '_'.join(x))
        for i,name in enumerate(range(0,len(data_1[DF3_column].unique()))):
            globals()['data_1_%s'%name]=data_1[data_1[DF3_column] ==data_1[DF3_column].unique()[i]]
        data_1[DF3_column]=data_1[DF3_column].apply(lambda x : ''.join(x))
        data_1[DF3_column]=data_1[DF3_column].apply(lambda x: x.split("_"))
        for i in range(0,data_1.shape[0]):
            data_1[DF3_column][i].sort()
        data_1[DF3_column]=data_1[DF3_column].apply(lambda x: '_'.join(x))
        for i,name in enumerate(data_1[DF3_column].unique()):
            globals()['data_1_%s'%name]=data_1[data_1[DF3_column] ==data_1[DF3_column].unique()[i]]
            globals()['data_1_%s'%name].to_csv(address+"/%s.csv"%name)
            
        

    def tocsv(self,data_,address):
        data_.to_csv(address)
        
   
```
